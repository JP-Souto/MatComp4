# <center><u>**1. Ondas**</u></center>

<br>

Na física, uma onda é uma perturbação que viaja pelo espaço e pela matéria através da transferência de energia de um ponto a outro. Modelos matemáticos que descrevem ondas são usados por engenheiros para tratar de sistemas que oscilam ou vibram. Neste contexto, as **funções trigonométricas** desempenham um papel fundamental.

Ondas frequentemente são usualmente tratadas como **funções periódicas** (ou sinais periódicos). Um sinal periódico é definido como aquele que apresenta repetição de seus valores em um intervalo definido de tempo (período). Ou seja, para qualquer $\theta\in\mathbb R$ temos

$$
f(\theta) = f(\theta + nT),
$$

onde $T$ é o valor (fixo) deste intervalo (chamado **período**) e $n\in\mathbb Z$ indica um número inteiro qualquer.

<br><br>

Considerando o ângulo $\theta$ como uma função linear do tempo $t$, tal que

$$\theta = \omega t,
$$

temos:
- $\theta=$ ângulo ($rad$),
- $\omega=$ frequência angular ($rad/s$),
- $t=$ tempo ($s$).

Para um intervalo $\Delta \theta$ temos

$$\Delta \theta = \omega \Delta t, \quad\quad \text{ou}  \quad\quad  \Delta t = \frac{\Delta \theta}{\omega}.$$

<br>

Assim, para a função completar um ciclo (2$\pi$), o tempo associado é definido como o **período**

$$T = \frac{\Delta \theta}{\omega} = \frac{2\pi}{\omega},$$

medido em segundos ($s$).

<br>

O inverso do período, representando a quantidade de ciclos por unidade de tempo, é denominado **frequência** (comum):

$$F = \frac{1}{T} = \frac{\omega}{2\pi},$$

medida em ciclos por segundo, ou hertz ($Hz$).

## **Funções senoidais**

<br>

Uma forma ondulatória geral é descrita por uma função seno ou cosseno, denominada senóide, que pode ser expressada como

$$ f(t) = C + A \sin(\omega t + \phi),$$

onde:
- $\phi$ é o chamado ângulo de fase,
- $A$ é a amplitude,
- $\omega$ é frequência angular,
- $C$ é uma translação vertical (ou valor médio), e
- o período é dado por $T = \frac{2\pi}{\omega}$.

<br>

O código abaixo mostra a construção do gráfico de uma senóide, mostrando o período e a amplitude.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 2.0
T = 2.0
phi = 0.5
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
y = A*np.sin(T*x+phi) #T multiplica por cada termo do array x

fig, ax = plt.subplots(figsize=(12, 2))

# desenhando a medida do periodo (comprimento)
ax.text(np.pi/T,A+0.6, "T", style='italic', size='x-large')
ax.hlines(y=A+0.4, xmin=0, xmax= 2*np.pi/T, color='k', lw=1.0)
ax.plot(0, A+0.4, 'k|')
ax.plot(2*np.pi/T, A+0.4, 'k|')

# desenhando a medida da amplitude
ax.text(-0.3,A/2, "$A$", style='italic', size='x-large')
ax.vlines(x= 0, ymin=0, ymax= A, color='k', lw=1.0)
ax.plot(0, 0, 'k_')
ax.plot(0, A, 'k_')

ax.set_ylim(-(A+1.5), A+1.5)
plt.plot(x,y)
plt.grid()
plt.show()
```

Dizemos que a curva acima é descrita no **domínio do tempo**.

Embora não seja tão familiar, o **dominio de frequência** fornece uma perspectiva alternativa para caracterizar o comportamento de funções oscilantes. Conhecendo a amplitude máxima positiva de uma função senoidal, a posição de $1/T$ no *eixo de frequências* e o ângulo de fase, tem-se informações suficientes para reproduzir a curva no domínio do tempo. A descrição de sinais no domínio da frequência é usada para caracterizar sinais mais complicadas usando a <u>**transformada de Fourier**</u>.

## **A onda quadrada**

Para construir uma onda quadrada em Python é conveniente utilizar uma sentença condicional que retorna $1$ (`True`) se for verdadeira ou $0$ se for falsa (`False`). Por exemplo, para avaliar quais valores inteiros entre $-5$ e $5$ são maiores que $1$ podemos fazer como é mostrado a seguir:

```python
x =  np.arange(-5, 5, 1)    # intervala os numeros das duas primeiras entradas e escolhe o espaçamento no parametro 3
print('Inteiros entre -5 e 5:      ', x)
print('Condicional maior que 1:    ', x > 1 )
print('Valor lógico correspondente:', np.int32(x > 1))
```

### **Exemplo 1:**
Para construir uma única onda quadrada definida pela função

\begin{equation}
f(t) = \begin {cases}
                 1, \ \ \text{se } -1<t<1   \\
                 0, \ \ \text{se } t\leq -1 \text{ ou } 1 \leq t \end{cases}
\end{equation}
no intervalo $-2\leq t \leq2$, podemos fazer como é mostrado a seguir:

```python
t = np.linspace(-2, 2, 1000)    # gerar valores (1000) entre um intervalo (-2 e 2)
plt.plot(t, (-1<t)&(t<1))       # comente
plt.grid()
plt.show()
```

<br>

***

**Observação:** As linhas verticais em $x=-1$ e $x=1$ não fazem parte do gráfico da função. Uma representação mais precisa deveria omitir essas linhas e usar a simbologia de intervalo fechado e aberto nesses pontos.

***

<br>

Para estender <u>periodicamente</u> a onda do exemplo anterior (de período $T=4$) ao longo de um intervalo maior, por exemplo $-6\leq t \leq 6$, podemos fazer como é mostrado no exemplo a seguir.

<br>

### **Exemplo 2:**
Estendendo a função $f=f(t)$ do exemplo anterior, definida inicialmente em $[-2, 2]$, para qualquer valor de $t\in\mathbb R$.

```python
# comente o código abaixo
def f2(x):
    if x<-2:
        return f2(x + 4) #4 é o valor do período(T)
    elif x>2:
        return f2(x - 4)
    elif (0<x) & (x<1) or (-1<x) & (x<0):
        return 1
    else:
        return 0

x = np.linspace(-6, 6, 1000)
plt.plot(x, [f2(xi) for xi in x])
plt.grid()
plt.show()
```

<br>

### **Exemplo 3:**
Gráfico da função $f(t)=2\cos(2\pi t)$, de amplitude $A=2$ e período $T=\frac{2\pi}{2\pi}=1$ pode ser gerado como é mostrado abaixo.

```python
t = np.linspace(-4, 4, 1000)
#f3 = lambda t: complete
f3 = lambda t: 2*np.cos(2*np.pi*t)

plt.plot(t, f3(t))
plt.grid()
```

A função `np.sign()` determina o sinal (positivo, negativo ou zero) dos elementos de um `array`. A função retorna um `array` com $1$ para os elementos positivos, $-1$ para os negativos e $0$ para os nulos. Assim, uma onda quadrada com período $T=1$ pode ser gerada a partir da função do exemplo anterior como é mostrado a seguir (mas note que sua amplitudo, por definição, será de $A=1$).


```python
plt.plot(t, np.sign(f3(t)))
plt.grid()
```

## <u>**Exercício 1:**</u>

Somando dois sinais diferentes podemos obter um sinal diferente, formado pela composição dos dois sinais  originais. Por exemplo, somando os sinais $f_1(t)$ e $f_2(t)$ obtemos o resultado mostrado a seguir.

```python
t = np.linspace(-4,4,1000)
f1 = lambda t: 2*np.sign(np.sin(2*np.pi*t))
f2 = lambda t: np.sign(np.sin(4*np.pi*t))
fig, ax = plt.subplots(figsize=(10,2))
ax.set_ylim(-4,5)
plt.plot(t, f1(t), "r", label = "$f_1$")
plt.legend()
plt.grid()

fig, ax = plt.subplots(figsize=(10,2))
ax.set_ylim(-4,5)
plt.plot(t, f2(t), "b", label = "$f_2$")
plt.legend()
plt.grid()

fig, ax = plt.subplots(figsize=(10,2))
ax.set_ylim(-4,5)
plt.plot(t, f1(t), "r--", label = "$f_1$", lw=0.7)
plt.plot(t, f2(t), "b--", label = "$f_2$", lw=0.7)
plt.plot(t, f1(t)+f2(t),"k",label = "$f_1+f_2$")
plt.legend()
plt.grid()
plt.show()
```

Com base no exemplo acima, obtenha o sinal da figura abaixo:

```python
#resolva

t = np.linspace(-4,4,1000)
f1 = lambda t: np.sign(np.sin(np.pi*t+np.pi/2))
f2 = lambda t: 2*np.sign(np.sin(2*np.pi*t+np.pi/2))

fig, ax = plt.subplots(figsize=(10,2))
ax.set_ylim(-4,5)
plt.plot(t, f1(t), "r--", label = "$f_1$", lw=0.7)
plt.plot(t, f2(t), "b--", label = "$f_2$", lw=0.7)
plt.plot(t, f1(t)+f2(t),"k",label = "$f_1+f_2$")
plt.legend()
plt.grid()
plt.show()
```
